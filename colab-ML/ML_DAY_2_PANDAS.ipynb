```python
v = pd.read_table("https://raw.githubusercontent.com/nunnarilabs/ml/master/pandas/chipotle.tsv")
v
```

```python
v.isnull().sum()
v.dropna(inplace=True)
```

```python
v['item_price']=v['item_price'].str.replace('$','')
```

```python
v.item_price = v.item_price.astype(float)
```

```python
type(v.item_price)
```

```python
v.item_price*100
v.item_price
```

```python
v.item_name.unique
```

```python
df_drink = v[v.item_price<1.1]
df_drink
```

```python
df_drink.item_name.nunique()
```

```python
sk = df_drink.item_name.value_counts()
sk
```

```python
w=v[v.item_name=='Steak Burrito'].choice_description.value_counts()
w
```
