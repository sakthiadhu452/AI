```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Global YouTube Statistics.csv', encoding='ISO-8859-1')
df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
df["lowest_yearly_earnings"].replace(np.nan,df['lowest_yearly_earnings'].astype('float').mean(axis=0),inplace=True)
df["highest_yearly_earnings"].replace(np.nan,df['highest_yearly_earnings'].astype('float').mean(axis=0),inplace=True)
df["Latitude"].replace(np.nan,df['Latitude'].astype('float').mean(axis=0),inplace=True)
df["subscribers"].replace(np.nan,df['subscribers'].astype('float').mean(axis=0),inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['subscribers'].unique()
```

```python
df.shape
```

```python
avg=df["subscribers"].astype("float").mean(axis=0)
print(avg)
```

```python
df["subscribers"].replace(np.nan,avg,inplace=True)
df
meanOfBore=df["uploads"].astype(float).mean(axis=0)
print(meanOfBore)
df["uploads"].replace(np.nan,meanOfBore,inplace=True)
df
```

```python
k=df["subscribers"].mode()
print(k)
```

```python
df["subscribers"].replace(np.nan,'k',inplace=True)
df
```

```python
df.dropna(subset=["uploads"],axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
from scipy import stats
hp_zscore=np.abs(stats.zscore(df['subscribers']))#z=x-miu/standard deviation
hp_zscore
```

```python
thre=hp_zscore.max()
index_val=np.where(hp_zscore>=thre)
df.iloc[index_val]['subscribers']
```

```python
nd=df.drop(index=df.iloc[index_val]['subscribers'].index)
nd
```
